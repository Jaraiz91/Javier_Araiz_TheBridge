### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline

import pandas as pd
import numpy as np
import matplotlib_inline
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
address = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(address, sep="\t")
chipo
```

### Step 3. See the first 10 entries

```python
chipo.head(10)
```

### Step 4. Create a histogram of the top 5 items bought

```python
count_items = chipo.groupby(chipo["item_name"])["quantity"].sum()
count_items = pd.DataFrame(count_items, columns= ["quantity"])
count_items = count_items.sort_values("quantity")
five_most = count_items.tail(5)
five_most
```

```python
five_most.plot(kind = 'bar')
```

### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
price = [float(x[1:-1]) for x in chipo.item_price]
chipo.item_price = price
price_items = chipo.groupby(chipo["order_id"])["item_price"].sum()
price_items = pd.DataFrame(price_items, columns= ["item_price"])
price_items
```

```python
prices = chipo.groupby(chipo['order_id']).sum()
plt.scatter(x = prices.item_price, y= prices.quantity,s= 50,  c= 'green')
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```
