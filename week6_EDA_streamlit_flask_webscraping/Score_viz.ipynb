### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline
import pandas as pd
import numpy as np
import matplotlib_inline
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
data = {"first_name": ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'age':[42, 52, 36, 24, 76], 'female': [0, 1, 1, 0, 1], 'preTestScore': [4, 24, 31, 2, 3], 'postTestScore': [25, 94, 57, 62, 70]}
```

```python
df = pd.DataFrame(data)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.scatterplot(data= df, x= "preTestScore", y= "postTestScore", size= "age")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.scatterplot(data= df, x= "preTestScore", y= "postTestScore", size=df['postTestScore'] * 4.5, sizes=(50, 300), hue= 'female', palette=("blue", "red"), legend = False)
```
